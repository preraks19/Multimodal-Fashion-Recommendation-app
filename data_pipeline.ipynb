```python
import numpy as np
import os, requests
import pandas as pd
```

```python
df_styles = pd.read_csv(os.path.join('data', 'styles.csv'), error_bad_lines=False)
df_styles = df_styles[df_styles['masterCategory'] == 'Apparel']
df_styles['id'] = df_styles['id'].astype(str) + '.jpg'
df_styles.head()
```

```python
df_images = pd.read_csv(os.path.join('data', 'images.csv'), error_bad_lines=False)
df_images.head()
```

```python
# randomly choose 1000 images
df_images = df_images.sample(n=1000)
# df_styles = df_styles[df_styles['id'].isin(df_images['filename'])]

# sample 1000 images common in both dataframes
df_styles = df_styles[df_styles['id'].isin(df_images['filename'])]
df_images = df_images[df_images['filename'].isin(df_styles['id'])]


df_images.to_csv(os.path.join('preprocessed_data', 'images.csv'), index=False)
df_styles.to_csv(os.path.join('preprocessed_data', 'styles.csv'), index=False)
```

```python
def download_image(url, filename):
    filename = os.path.join('images', filename)
    r = requests.get(url, allow_redirects=True)
    open(filename, 'wb').write(r.content)
```

```python
# download images
for url, filename in zip(df_images['link'], df_images['filename']):
    download_image(url, filename)
```
